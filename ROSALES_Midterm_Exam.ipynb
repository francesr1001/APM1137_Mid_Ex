# Langrange





<section id="sectionlang"> </section>


```python
# VALUES

import numpy as np
import math

def lagrange_interpolation(x, y, xi, degree):
    n = len(x)
    result = 0
    for d in range(degree + 1):
        term = y[d]
        for j in range(n):
            if j != d:
                term *= (xi - x[j]) / (x[d] - x[j])
        result += term
    return result

degree = int(input("Enter the degree of Lagrange polynomial to find (1st, 2nd, 3rd, etc.): "))
num_nodes = degree + 1
print(f"You need to input {num_nodes} nodes.")

function_string = input("Enter the function f(x): ")

nodes = []
function_values = []
for i in range(num_nodes):
    node = float(input(f"Enter node {i+1}: "))
    nodes.append(node)
    function_values.append(eval(function_string.replace('x', str(node)), {"sin": math.sin, "cos": math.cos}))

lagrange_polynomials = []
for i in range(num_nodes):
    polynomial = ""
    for j in range(num_nodes):
        if j != i:
            polynomial += f"(x - {nodes[j]}) / ({nodes[i]} - {nodes[j]}) * "
    polynomial = polynomial[:-3]  
    lagrange_polynomials.append(polynomial)

lagrange_sum = ""
for i in range(num_nodes):
    lagrange_sum += f"f({nodes[i]}) * ({lagrange_polynomials[i]}) + "

lagrange_sum = lagrange_sum[:-3]

for i in range(num_nodes):
    lagrange_sum = lagrange_sum.replace(f"f({nodes[i]})", str(function_values[i]))

print("The Lagrange interpolation result is:")
print(lagrange_sum)


print("Input values:")
for i in range(num_nodes):
    print(f"Node {i+1}: {nodes[i]}, f({nodes[i]}): {function_values[i]}")
```

#  Divided Difference






<section id="sectiondiff"> </section>

```python
# VALUES

def divided_difference(x, y):
    n = len(y)
    coefficients = []
    for i in range(n):
        coefficients.append(y[i])
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            coefficients[i] = (coefficients[i] - coefficients[i-1]) / (x[i] - x[i-j])
    return coefficients

def interpolated_polynomial(x, y):
    n = len(x)
    coefficients = divided_difference(x, y)
    polynomial = f"{coefficients[0]}"
    for i in range(1, n):
        polynomial += ''.join([f"{'+' if coefficients[i] >= 0 else ''} {coefficients[i]}"])
        for j in range(i):
            polynomial += f"(x - {x[j]})"
    return polynomial

def main():
    num_points = int(input("Enter the number of function points: "))
    x_values = []
    y_values = []
    print("Enter the x and y values for each function point:")
    for i in range(num_points):
        x = float(input(f"Enter x{i}: "))
        y = float(input(f"Enter y{i}: "))
        x_values.append(x)
        y_values.append(y)

    polynomial = interpolated_polynomial(x_values, y_values)
    print("\nYou entered the following function points:")
    for i in range(num_points):
        print(f"x{i}: {x_values[i]}, y{i}: {y_values[i]}")

    print("\nInterpolated polynomial:")
    print(polynomial)

if __name__ == "__main__":
    main()
```

# Spine Interpolation






<section id="sectionspline"> </section>

```python
# VALUES

def linear_spline_interpolation(nodes):
    s = []
    for i in range(len(nodes) - 1):
        x0, y0 = nodes[i]
        x1, y1 = nodes[i + 1]
        slope = (y1 - y0) / (x1 - x0)
        intercept = y0 - slope * x0
        s.append((slope, intercept))
    return s

def quadratic_spline_interpolation(nodes):
    s = []
    for i in range(len(nodes) - 1):
        x0, y0 = nodes[i]
        x1, y1 = nodes[i + 1]
        slope0 = (y1 - y0) / (x1 - x0)
        slope1 = slope0
        intercept0 = y0 - slope0 * x0
        intercept1 = y1 - slope1 * x1
        s.append((intercept0, slope0, slope1))
    return s

def main():
    degree = int(input("Enter the degree of interpolation (1 for linear, 2 for quadratic): "))
    num_pairs = int(input("Enter the number of pairs of nodes: "))

    nodes = []
    print("Enter the x and y values for each pair of nodes:")
    for i in range(num_pairs):
        x = float(input(f"Enter x{i + 1}: "))
        y = float(input(f"Enter y{i + 1}: "))
        nodes.append((x, y))

    print("\nYou entered the following pairs of nodes:")
    for i, node in enumerate(nodes):
        print(f"x{i + 1} = {node[0]}, y{i + 1} = {node[1]}")

    if degree == 1:
        splines = linear_spline_interpolation(nodes)
        print("\nLinear spline interpolation:")
        for i, spline in enumerate(splines):
            x0, _ = nodes[i]
            x1, _ = nodes[i + 1]
            print(f"For x in [{x0}, {x1}]:")
            print(f"s_{i + 1}(x) = {spline[0]} * x + {spline[1]}")
    elif degree == 2:
        splines = quadratic_spline_interpolation(nodes)
        print("\nQuadratic spline interpolation:")
        for i, spline in enumerate(splines):
            x0, _ = nodes[i]
            x1, _ = nodes[i + 1]
            print(f"For x in [{x0}, {x1}]:")
            print(f"s_{i + 1}(x) = {spline[0]} + {spline[1]} * x + {spline[2]} * x^2")

if __name__ == "__main__":
    main()
```
